```python
def find_foreign(start: int, end: int, sentence: str):
    count=0
    a = int(start, 16) 
    b = int(end, 16)
    return_sentence = ''
    for i, w in enumerate(sentence):
        if a <= ord(w) and ord(w) <= b:
            return True
    return False
```

```python
from koeda import RD
from koeda import RI
from koeda import SR
from koeda import RS
sen="2000년에 청소년 야구 국가대표팀으로 뽑혀 세계 청소년 야구 국제대회에 출전해 동갑인 이대호, 정근우, 추신수와 함께 청소년 대표팀 우승을 이끌었다."
rd = RD(morpheme_analyzer="Okt")
rd_result=(rd(sen))
ri = RI(morpheme_analyzer="Okt")
ri_result=(ri(sen))
sr = SR(morpheme_analyzer="Okt")
sr_result=(sr(sen))
rs = RS(morpheme_analyzer="Okt")
rs_result=(rs(sen))

print(rd_result)
print(ri_result)
print(sr_result)
print(rs_result)
```

```python
import pandas as pd
import re
df = pd.read_csv('input.csv')
df
```

```python
def get_only_hangul(line):
	parseText= re.compile('/ ^[ㄱ-ㅎㅏ-ㅣ가-힣]*$/').sub('',line)

	return parseText
```

```python
trans_list=[]
for i, sen, sub, obj, label, source in notebook.tqdm(zip(df['id'], df['sentence'], df['subject_entity'], df['object_entity'], df['label'], df['source'])):
    #no_relation이 아니면서 / 외국어가 전혀 없으면서(영어포함) / 각 엔티티가 문장에 한 번만 존재하는 경우에
    if( label!="no_relation"):
        if(find_foreign('4e00', '9fff', sen)==False) and (find_foreign('0400', '04ff', sen)==False) and (find_foreign('0600', '06ff', sen)==False) and (find_foreign('0370', '03ff', sen)==False) and (find_foreign('3040', '309f', sen)==False) and (find_foreign('30a0', '30ff', sen)==False):
            sub = eval(sub)
            obj = eval(obj)
    
            sub_word=sub['word']
            obj_word=obj['word']
    
            sub_word_cnt=sen.count(sub_word)
            obj_word_cnt=sen.count(obj_word)
            
            if(sub_word_cnt==1 and obj_word_cnt==1):
                hangul_sen=get_only_hangul(sen) #영어 제거
                new_sen=sr(hangul_sen) #eda 지정(rd, ri, rs, sr)
                if sub_word in new_sen and obj_word in new_sen:#번역시에도 문장의 엔티티가 잘 살아있는 경우에만 추가
                    if(new_sen!=sen): #같은 문장이 돌아오는 경우가 있어 문장이 다를 경우에만
                        ss=new_sen.find(sub_word)
                        se=ss+len(sub_word)-1
                        os=new_sen.find(obj_word)
                        oe=os+len(obj_word)-1
                        
                        sub['start_idx']=ss
                        sub['end_idx']=se
                        obj['start_idx']=os
                        obj['end_idx']=oe
                        
                        sub=str(sub)
                        obj=str(obj)
                        
                        temp=[]
                        temp.append(i)
                        temp.append(new_sen)
                        temp.append(sub)
                        temp.append(obj)
                        temp.append(label)
                        temp.append(source)
                        trans_list.append(temp)
```

```python
new_df = pd.DataFrame(trans_list)
new_df
new_df.to_csv('output_sr.csv',index=False)
```
