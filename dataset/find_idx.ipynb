### 아무것도 출력이 되지 않으면, idx가 바르게 되어있다는 것임

```python
for sen, sub, obj in zip(test_df['sentence'], test_df['subject_entity'], test_df['object_entity']):
    raw_sub = eval(sub)
    raw_obj = eval(obj)
    ss=raw_sub['start_idx']
    se=raw_sub['end_idx']
    os=raw_obj['start_idx']
    oe=raw_obj['end_idx']
    if(sen[ss:se+1]!=raw_sub['word']):
        print(sen[ss:se+1])
        print(raw_sub['word'])
        
    if(sen[os:oe+1]!=raw_obj['word']):
        print(sen[os:oe+1])
        print(raw_obj['word'])
```

```python
star="★★★"
def find_idx(sen):
    iterator = re.finditer(star ,sen)
    save=[]
    for match in iterator:
        temp=[]
        temp.append(match.start())
        temp.append(match.end())
        save.append(temp)
    return save
```

```python
idx=0
for sen, sub, obj in zip(df['sentence'], df['subject_entity'], df['object_entity']):
    raw_sub = eval(sub)
    raw_obj = eval(obj)
    old_ss=raw_sub['start_idx']
    old_se=raw_sub['end_idx']
    old_os=raw_obj['start_idx']
    old_oe=raw_obj['end_idx']
    
    sub_word=raw_sub['word']
    obj_word=raw_obj['word']
    idx_info=find_idx(sen)
    
    #그러나 남편 @@@콘라트 2세@@@가 황제관을 쓴 뒤 13일 만에 @@@기젤라@@@는 황후로 책봉되었다.
    #print(sen)
    #별 제거전
    star1_s = idx_info[0][1]
    star1_e = idx_info[1][0]
    star2_s = idx_info[2][1]
    star2_e = idx_info[3][0]

    if(sen[star1_s:star1_e]==sub_word): #subect_entity idx가 더 작음
        ss=star1_s-3
        se=star1_e-4
        os=star2_s-9
        oe=star2_e-10
    else:
        os=star1_s-3
        oe=star1_e-4
        ss=star2_s-9
        se=star2_e-10
    
    no_star_sen=sen.replace("★★★","")
    raw_sub['start_idx']=ss
    raw_sub['end_idx']=se
    raw_obj['start_idx']=os
    raw_obj['end_idx']=oe
    
    df['sentence'][idx]=no_star_sen
    df['subject_entity'][idx]=str(raw_sub)
    df['object_entity'][idx]=str(raw_obj)
    idx+=1
```
