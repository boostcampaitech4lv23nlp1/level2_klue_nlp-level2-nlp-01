### subject_entity, object_entity에 변화가 생겼으면 출력

```python
idx_before=[]
for sen, sub, obj in zip(df['sentence'], df['subject_entity'], df['object_entity']):
    temp=[]
    raw_sub = eval(sub)
    raw_obj = eval(obj)
    sub_word=raw_sub['word']
    obj_word=raw_obj['word']
    temp.append(sub_word)
    temp.append(obj_word)
    idx_before.append(temp)
```

```python
idx_after=[]
for sen, sub, obj in zip(df1['sentence'], df1['subject_entity'], df1['object_entity']):
    temp=[]
    raw_sub = eval(sub)
    raw_obj = eval(obj)
    sub_word=raw_sub['word']
    obj_word=raw_obj['word']
    temp.append(sub_word)
    temp.append(obj_word)
    idx_after.append(temp)
```

```python
for i,j in zip(idx_before,idx_after):
    if(i!=j):
        print(i)
        print(j)
```

### 인덱스에 문제가 있으면 출력

```python
for sen, sub, obj in zip(df['sentence'], df['subject_entity'], df['object_entity']):
    raw_sub = eval(sub)
    raw_obj = eval(obj)
    ss=raw_sub['start_idx']
    se=raw_sub['end_idx']
    os=raw_obj['start_idx']
    oe=raw_obj['end_idx']
    if(sen[ss:se+1]!=raw_sub['word']):
        print(sen[ss:se+1])
        print(raw_sub['word'])
        
    if(sen[os:oe+1]!=raw_obj['word']):
        print(sen[os:oe+1])
        print(raw_obj['word'])
```
